```python
# Find all occurrences of a pattern in a string
# Input: Two strings, Pattern and Genome.
# Output: All starting positions where Pattern appears as a substring of Genome.

def findOccur(pattern, genome):
    result = []
    for i in xrange(0, len(genome) - len(pattern) + 1):
        subString = genome[i:i+len(pattern)]
        if pattern == subString:
            result.append(i)
    return result
```

```python
def freqeutWords2(leastCount , text, k):
    """
    text: string, text info
    k: int, requent pattern that to be search
    return:
        the most frequent patterns
    """
    patterns = {}
    mostCount = 0
    result = set()
    for i in xrange(0, len(text) - k + 1):
        subString = text[i:i+k]
        if subString in patterns:
            patterns[subString] += 1
        else:
            patterns[subString] = 1
        if patterns[subString] > mostCount:
            mostCount = patterns[subString]
    for key in patterns:
        #if patterns[key] == mostCount:
        if patterns[key] >= leastCount:
            #print key,
            result.add(key)
    return result, mostCount

def findClump(genome, k, clump, leastCount):
    subGenome = []
    i = 0
    result = set()
    while (i+1) * clump < len(genome):
        subGenome.append(genome[i * clump: (i+1)*clump])
        i += 1
    subGenome.append(genome[i * clump: -1])
    for seq in subGenome:
        patterns, mostcount = freqeutWords2(leastCount , seq, k)
        for item in patterns:
            result.add(item)
    return result

def findClumpEveryWindow(genome, k, clump, leastCount):
    result = set()
    for i in xrange(len(genome) - clump + 1):
        subGenome = genome[i:i+clump]
        patterns,mostCount = freqeutWords2(leastCount, subGenome, k);
        for item in patterns:
            result.add(item)
    return result
```

```python
#Approximate Pattern Matching Problem: Find all approximate occurrences of a pattern in a string.
#Input: Strings Pattern and Text along with an integer d.
#Output: All starting positions where Pattern appears as a substring of Text with at most d mismatches
```

```python
def approxPattern(str1, template, distance):
    result = []
    for i in xrange(len(template) - len(str1) + 1):
        subString = template[i:i+len(str1)]
        if hammingDistance(subString, str1) <= distance:
            result.append(i)
    return result
```

```python
# Cheat version:
# first calculate all the exact match string, then use exact match string to compare
# 
def findMismatch(string, length, misValue):
    toCheckStrings = set()
    for i in xrange(len(string) - length + 1):
        toCheckStrings.add(string[i:i+length])
    values = dict()
    for item in toCheckStrings:
        values[item] = 0
    for i in xrange(len(string) - length + 1):
        for item in values:
            if hammingDistance(item, string[i:i+length]) <= misValue:
                values[item] += 1
    return values
```

```python
# None Cheat version:
# first calculate all the exact match string, then use exact match string to compare
# 
from Bio.Seq import Seq
def findMismatchComp(string, length, misValue):
    container = set()
    container.add("A")
    container.add("T")
    container.add("G")
    container.add("C")
    for i in xrange(length-1):
        newContainer = set()
        for item in container:
            newContainer.add(item + "T")
            newContainer.add(item + "C")
            newContainer.add(item + "G")
            newContainer.add(item + "A")
        container = newContainer

    toCheckStrings = container;
    #for i in xrange(len(string) - length + 1):
        #toCheckStrings.add(string[i:i+length])
    values = dict()
    result = dict()
    for item in toCheckStrings:
        values[item] = 0
        result[item] = 0
    #result = values
    for i in xrange(len(string) - length + 1):
        for item in values:
            if hammingDistance(item, string[i:i+length]) <= misValue:
                values[item] += 1
    for item in values:
        test = Seq(item)
        result[item] = values[item] + values[test.reverse_complement().tostring()]
    return result, values
```
